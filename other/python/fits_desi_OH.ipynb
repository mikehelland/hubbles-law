```python
import math
import fitsio

columns = ['O_ABUNDANCE', 'Z']
d = fitsio.read('Fig10left_data.fits', columns=columns)
d2 = fitsio.read('Fig13_data.fits', columns=columns)
```

```python
f = open("xmpg", "a")

for i in d:
    try:
        f.write('{"z":' + str(round(i[1], 4)) + ',"OH":' + str(round(i[0], 4)) + '},')
        
        
    except:
        print("skip")
    
f.close()
```

```python
import matplotlib.pyplot as plt

dataz = [row[1] for row in d] 
dataO = [row[0] for row in d]

dataz2 = [row[1] for row in d2] 
dataO2 = [row[0] for row in d2]

plt.plot(dataz2, dataO2, 'bo', label = "oxygen emitters")
plt.plot(dataz, dataO, 'ro', label = "xmpg")

plt.xlabel("Redshift (z)")
plt.ylabel("12+log(O/H)")
plt.legend()

plt.savefig('o-xmpg.png',  bbox_inches='tight')
plt.show()
```
