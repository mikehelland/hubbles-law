```python
data = [0]

H = 70
H = H / 3.08e19 * 60 * 60 * 24 * 365 * 1e6
c = 1
z = 0
t = 0
x = 0

while x < 7000:
    t -= 1
    x += c / (1 + z)
    z = c / (c - H * x) - 1

    data.append(x)

```

```python
import matplotlib.pyplot as plt

plt.style.use("dark_background")
dataX = []
dataT = []


for i in range(len(data)):
    dataT.append(-i / 1000)
    dataX.append(data[i] / 1000)

plt.plot(dataX, dataT, label = "photon", color = "yellow")
    
plt.legend()
plt.savefig('tdpmatplot.png',  bbox_inches='tight')
plt.show()
```
