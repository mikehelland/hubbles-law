```python
H0km = 70
ΩΛ = 0.7
Ωm = 0.3

H0 = H0km / 3.08e19 * 60*60*24*365 * 1e6

c = 1
a = 1
t = 0
x = 0

data = []

while x >= 0:
    H = H0 * (Ωm * a**-3 + ΩΛ)**0.5
    
    t -= 1
    x += c - H * x    
    a -= H * a
        
    data.append([t, x])
```

```python
import matplotlib.pyplot as plt

plt.axis([0, 14, -14, 0])
#plt.style.use("dark_background")

plt.plot([row[1]/1000 for row in data], [row[0]/1000 for row in data])
#color = "yellow"

plt.title("$H_0 = " + str(H0km) + "$ km s$^{-1}$ Mpc$^{-1}$, $\Omega_\Lambda =" + str(ΩΛ) + ", \Omega_m = " + str(Ωm) + "$")

plt.xlabel("Proper distance (billion light years)")
plt.ylabel("Time (billion years)")
#plt.legend()

plt.savefig('photon_simple.png',  bbox_inches='tight')
plt.show()
```
